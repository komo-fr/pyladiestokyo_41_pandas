# 最初の一歩

```python
# 2018/1/1から365日分のデータをランダムに生成 (標準偏差1の正規分布に従う乱数)
ts = pd.Series(np.random.randn(365),
                      index=pd.date_range('1/1/2018', periods=365))
ts.head(10)
```

```python
ts.describe()
```

```python
ts.plot()
```

```python
# 累積を求めてみる
tsc = ts.cumsum()
tsc.head()
```

```python
tsc.plot()
```

```python
df = pd.DataFrame(np.random.randn(365, 4),
                             index=ts.index, columns=list('ABCD'))
```

```python
df.head()
```

```python
df.describe()
```

```python
plt.figure()
df.plot()
```

```python
cumsum_df = df.cumsum()
cumsum_df.head()
```

```python
cumsum_df.plot()
```

- `plot`を使うと、手軽にグラフを描画できる
- `plot`に`kind`を指定すると、グラフの種類を指定できる
    - ‘bar’ / ‘barh’ 
    - ‘hist’ 
    - ‘box’ 
    - ‘kde’ / ‘density’ 
    - ‘area’ 
    - ‘scatter’ 
    - ‘hexbin’ 
    - ‘pie’

## ヒストグラム

```python
iris_df.plot(kind='hist')
```

```python
# 同じものが表示される
iris_df.plot.hist()
```

- plot.hist()とplot(kind='hist')は同じ
- kind='hist'は、pandasの0.17.0で追加された記述方法

```python
# 特定の列だけを表示
iris_df['sepal_length'].plot.hist()
```

```python
# ヒストグラムの幅を指定
iris_df['sepal_length'].plot.hist(bins=20)
```

```python
# 項目ごとにサブプロッとに分けたい
iris_df.plot(kind='hist',
                subplots=True, # 項目ごとに分割する
                figsize=(10, 5), # 描画する面積
                layout=(2, 2), # レイアウト（行数, 列数）
                bins=20)
```

```python
iris_df.hist()
```

- ヒストグラムの場合、plot.hist()やplot(kind='hist')以外に、hist()を使う方法もある
- https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.hist.html#pandas.DataFrame.hist

### speciesごとにプロット

```python
iris_df.groupby('species').size()
```

```python
iris_df.groupby('species').max()
```

```python
iris_df.groupby('species').describe().T
```

```python
# speciesごとにsepal_lengthのヒストグラムを描画
iris_df.groupby('species')['sepal_length'].plot(kind='hist')
```

```python
# speciesごとにsepal_lengthのヒストグラムを描画
iris_df.groupby('species')['sepal_length'].plot(kind='hist',
                                                                  legend=True,  # 凡例を表示
                                                                  alpha=0.7,  # 透明度を指定
                                                                  bins=20,
                                                                  figsize=(10, 5))
```

```python
iris_df.hist(by='species', column='sepal_width')
```

## 箱ひげ図

```python
iris_df.plot(kind='box')
```

```python
iris_df.boxplot(by='species', figsize=(20, 10), layout=(1, 4))
```

## 散布図

```python
iris_df.plot(kind='scatter', x="sepal_length", y="sepal_width")
```

```python
color_map = dict(setosa='red', versicolor='blue', virginica='green')
color_iris_df = iris_df.copy()
color_iris_df['color'] = iris_df['species'].apply(lambda x: color_map[x])
```

```python
color_iris_df.plot(kind='scatter',
                          x="sepal_length",
                          y="sepal_width",
                          c=color_iris_df['color'])
```

```python
iris_df.groupby('species')['sepal_width'].describe()
```

```python
color_iris_df.plot(kind='scatter',
                          x="sepal_length",
                          y="sepal_width",
                          s=color_iris_df['petal_width'] * 40,  # petal_widthを●の大きさに反映
                          figsize=(10, 5),
                          c=color_iris_df['color'],
                          alpha=0.7)
```

# 注意点
- 書き方が色々ある
   - ヒストグラム一つをとっても
       - df.plot()でkind引数に'hist'を指定する方法
       - df.plot.hist()を使う方法
       - df.hist()を使う方法
   - ドキュメントを読むときに、混乱しないようにしよう
- pandasを使えば簡単にグラフをかけるが、複雑なことをやろうとするとmatplotlibなど他のライブラリを使った方が楽なこともある
    - pandasのplotting機能の売りは、「簡単なグラフを簡単にかけること」
